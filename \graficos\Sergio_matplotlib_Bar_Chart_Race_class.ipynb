# Let's make a bar chart race for Brazil's COVID cases!
## - Let's do it by state!
## - Let's do it by month!


# - Step one: import/install required libraries

```
!pip install bar_chart_race
```

```
import pandas as pd
import bar_chart_race as bcr
import requests
```

# Step 2: download required data

```
r = requests.get('https://data.brasil.io/dataset/covid19/caso.csv.gz')
  
```

```
r.status_code
```

```
with open('caso.csv.gz','wb') as x :
  x.write(r.content)
  
```

```
df_cases = pd.read_csv('caso.csv.gz')
```

# Step 3: inspect the data
- Which columns do we want?

```
df_cases.head()
```

```
df_cases.select_dtypes(include = 'O')
```

```
df_cases.select_dtypes(include = ['number','bool'])
```

# Select the desired data
### - first, get the desired columns

```
df_cases_state = df_cases[['date','state','place_type','confirmed_per_100k_inhabitants']].copy()
df_cases_state
```

### - Then, filter the desired lines

```
filter = df_cases_state.place_type == 'state'
df_cases_state = df_cases_state.loc[filter]
df_cases_state
```

```
df_cases_state.place_type.unique()
```

```
# Step 3: Pivot Table!
```

### - Finally, convert the columns to the correct form

```
df_cases_state.dtypes
```

```
df_cases_state['date'] = pd.to_datetime(df_cases_state['date'])
df_cases_state.dtypes
```

# Step 3: get a column with the months

```
df_cases_state['date'].dt.month_name()
```

```
df_cases_state['year_month'] = '2020_' + df_cases_state['date'].dt.month.astype('str')
df_cases_state
```

# Step 4: pivot the table!
- lines: year_month
- columns: states
- values: confirmed
- aggregation: max

```
df_monthly_cases_pivoted = df_cases_state.pivot_table(index = 'year_month', columns = 'state', values = 'confirmed_per_100k_inhabitants', aggfunc = 'max')
df_monthly_cases_pivoted = df_monthly_cases_pivoted.fillna(0).astype(int)
df_monthly_cases_pivoted
```

```
bcr.bar_chart_race(
    df=df_monthly_cases_pivoted,
    filename='covid19_horiz.mp4',
    orientation='h',
    sort='desc',
    # n_bars=6,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=180,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    # period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total de casos: {v.nlargest(26).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    perpendicular_bar_func='median',
    period_length=1000,
    figsize=(5, 3),
    dpi=144,
    cmap='dark12',
    title='COVID-19 - casos por 100 mil habitantes',
    title_size='',
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family' : 'Helvetica', 'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False)  
```
