# PART 2: BASIC STATISTICS:

## PLEASE RUN THE CELLS BELOW, WITHOUT READING THEM, UNTIL PART 2.1!
#### (or stick around and see how I created the dataframe)



````python

# Let's generate a simple dataframe:
# generate lists of numbers:

np.random.seed(42) # all generated numbers will be the same 

dic = {} # dictionary where to store the data

# let's create lists with ages, sizes and weights
age = np.random.normal(loc = 30, scale = 5, size = 10) #normally generate random age distribution
size = np.random.normal(loc = 1.7, scale = 0.1, size = 10) #normally generate random size distribution (in meter)
weight = np.random.normal(loc = 80, scale = 10, size = 10) #normally generate random weight distribution (in kg)
gender = np.random.choice(['F','M'], 10) # randomly select genders to each person

# check out one of these lists!
weight
````

````python

# let's add these lists to a dictionary
dic['age'] = [int(x) for x in age] # convert all ages from float to int
dic['size'] = l = [round(x,2)  for x in size] # round all sizes
dic['weight'] = [int(x)   for x in weight] # convert all weights from float to int
dic['gender'] = gender

#check out the dictionary!
dic
````

````python

# let's create our dataframe
df = pd.DataFrame(dic)

# let's add a missing value!
df.loc[9,'gender'] = np.nan
df.head(2)
````

# PART 4 - HOW TO SPLIT DATA
## 4.1 - loc[]

- in a nutshell: insert inside the [] a dataframe that returns a boolean series
- this series of boolean acts as as filter for the dataframe index



````python
# ex of a boolean series: all men
df.gender == 'M'
````

````python
# how to bring only men
df[df.gender == 'M']
````

````python
# Let's make the code 'pythonic':
# filter: only men
fm = df.gender == 'M' # the filter
df[fm]
````

#### \\\#### WARNING: A CHALLENGE APPROACHES #####\\

````python
# filter: only female
ff = df.gender == 'F' # the filter
df[ff]
````

# BONUS: COMBINING SPLIT DATA WITH SCATTER PLOT!

````python
ax = df[fm].plot.scatter(x = 'age',y = 'weight', c = 'blue')
df[ff].plot.scatter(x = 'age',y = 'weight', c = 'red', ax = ax)
````

````python
f1 = df.gender == 'M'
f2 = df.weight < 74
df[f1 & f2]
````
